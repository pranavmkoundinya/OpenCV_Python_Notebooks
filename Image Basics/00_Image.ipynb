# Image

An image is a visual representation of an object, scene, person, or concept. Images can be captured, created, or rendered in various forms and through various means.

![image_intro](/home/pranavmkoundinya/Desktop/SPML/OpenCV_Python_Notebooks/stock_images/lighthouse_2.png)

Mathematically, an image is a function $f:\mathbb{R}\times\mathbb{R} \rightarrow \mathbb{R}$ of two spatial variables $x$ and $y$, where $f(x, y)$ represents the intensity at a spatial location $(x, y)$

# Digital Image

A digital image is a discretized representation of a visual object or a scene in a format that can be processed by a computer. Mathematically, it is a function $f:[1:M]\times[1:N]\rightarrow \mathbb{V}$ of two discrete spatial variables $i$ and $j$, where $f[i, j]$ represents the discretized intensity value at the spatial location $[i, j]$. Here $\mathbb{V}$ is a discretized set, and $M$, $N \in \mathbb{N}$. The discretized spatial location, is called a picture element, or **pixel** for short.

An image can be represented as a matrix, where each element corresponds to a pixel's intensity value

$$ I =
\begin{bmatrix}
I_{11} & I_{12} & . & . & I_{1N}\\
I_{21} & I_{22} & . & . & I_{2N}\\
. &  &  &  & .\\
. & &  &  & .\\
I_{M1} & I_{M2} & . & . & I_{MN}\\
\end{bmatrix}
$$

## Color Image

A color image is a digital image that includes color information for each pixel, allowing it to represent a wide range of colors. Unlike grayscale images, which only convey intensity information, color images use multiple color channels to capture and display colors. In an RGB color image, each pixel consist of a triplet of values, one value for each of red, green and blue channels.

![rgb](/home/pranavmkoundinya/Desktop/SPML/OpenCV_Python_Notebooks/media/rgb.jpeg)

```python
img_color = cv2.imread("/home/pranavmkoundinya/Desktop/SPML/OpenCV_Python_Notebooks/stock_images/garden.png")
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
plt.imshow(img_color)
```

Now, let us split the image into into its 3 constituent channels : R, G and B 

```python
# Red channel
r = img_color[:,:,0]
red_c = np.stack([r, np.zeros_like(r), np.zeros_like(r)], axis=2)

# Green channel
g = img_color[:,:,1]
green_c = np.stack([np.zeros_like(g), g, np.zeros_like(g)], axis=2)

# Blue channel
b = img_color[:,:,2]
blue_c = np.stack([np.zeros_like(b), np.zeros_like(b), b], axis=2)

fig = plt.figure(figsize=(20, 5))
fig.add_subplot(1, 4, 1)
plt.imshow(img_color)
fig.add_subplot(1, 4, 2)
plt.imshow(red_c)
fig.add_subplot(1, 4, 3)
plt.imshow(green_c)
fig.add_subplot(1, 4, 4)
plt.imshow(blue_c)
plt.show()
```
